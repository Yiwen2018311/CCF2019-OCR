# 关键地址切取

```python
import jieba
from gensim import corpora, models, similarities
```

```python
def remove_unuse_char(address):
    '''
    删除无用的行政区后缀
    村委会 ->村
    村村委会 -> 村
    村民委员会 - > 村
    村村民委员会 - > 村

    社区居民委员会 -> 社区
    社区居委会 -> 社区
    居委会 -> ''
    :param address:
    :return:
    '''
    address = address.replace('村村民委员会','村')
    address = address.replace('村民委员会', '村')
    address = address.replace('村村委会', '村')
    address = address.replace('村委会', '村')

    address = address.replace('社区居民委员会','社区')
    address = address.replace('社区居委会', '社区')
    address = address.replace('居委会', '')
    return address
```

```python
all_address_pd = pd.read_csv('../ext_data/nbs_id_card_code_ver1.csv')
```

```python
all_address_pd['address'] = all_address_pd.loc[:,'address'].apply(lambda x : remove_unuse_char(x))
```

```python
all_list =  all_address_pd.loc[:,'address']
```

```python
all_doc_list = []
for addr in all_list:
    doc_list = [word for word in jieba.cut_for_search(addr)]
#     doc_list = [word for word in jieba.cut(addr)]
    all_doc_list.append(doc_list)
```

```python
dictionary = corpora.Dictionary(all_doc_list)
```

```python
corpus = [dictionary.doc2bow(doc) for doc in all_doc_list]
```

```python
tfidf = models.TfidfModel(corpus)
```

```python
index = similarities.SparseMatrixSimilarity(tfidf[corpus], num_features=len(dictionary.keys()))
```

# 纠错开始

```python
correct_pd = result.copy()
for idx,error_id in enumerate(error_file_id):
#     print(error_id[0])
#     print(result.loc[result.file_id == error_id[0], 'addr'].values[0])

    doc_test= result.loc[result.file_id == error_id[0], 'addr'].values[0]
    doc_test_list = [word for word in jieba.cut_for_search(doc_test)]
    doc_test_vec = dictionary.doc2bow(doc_test_list)
    sim = index[tfidf[doc_test_vec]]
    
    similarity_index = sorted(enumerate(sim), key=lambda item: -item[1])[0][0]
    new_addr = remove_unuse_char(all_address_pd.iloc[similarity_index,12])
    
    print(idx,error_id[0],doc_test,'-->>',new_addr)
    correct_pd.loc[correct_pd.file_id == error_id[0], 'addr'] = new_addr
```

```python
output_path = '../single_result/1001_address_only_368_decode_ctc_l2_correct.csv'
```

```python
correct_pd.to_csv(output_path, encoding='utf_8_sig', index=False, header=None)
```
