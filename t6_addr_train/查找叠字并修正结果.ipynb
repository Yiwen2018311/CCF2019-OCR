```python
import pandas as pd
```

```python
result = pd.read_csv('../single_result/1001_address_only_368_backup.csv',header=None,names=['file_id','addr'])
```

```python
result[result.addr.str.len()>12].head()
```

```python
for idx,row in result[result.addr.str.len()>12].iterrows():
#     print(row['addr'])
    last_char = ''
    new_result = []
    for char in row['addr']:
        if last_char != char:
            new_result.extend(char)
            last_char = char
        else:
            last_char = ""
    result.iloc[idx,1] = "".join(new_result)
    print(row['file_id'],row['addr']," -> " ,"".join(new_result))
```
