# 关键地址切取

```python
import jieba
from gensim import corpora, models, similarities
```

```python
def remove_unuse_char(address):
    '''
    删除无用的行政区后缀
    村委会 ->村
    村村委会 -> 村
    村民委员会 - > 村
    村村民委员会 - > 村

    社区居民委员会 -> 社区
    社区居委会 -> 社区
    居委会 -> ''
    :param address:
    :return:
    '''
    address = address.replace('村村民委员会','村')
    address = address.replace('村民委员会', '村')
    address = address.replace('村村委会', '村')
    address = address.replace('村委会', '村')

    address = address.replace('社区居民委员会','社区')
    address = address.replace('社区居委会', '社区')
    address = address.replace('居委会', '')
    return address
```

```python
address_list = pd.read_csv('../ext_data/corpus/nbs_address_word_entry_for_train.txt',header=None).iloc[:, 0].tolist()
```

```python
# 添加专用词语 #best
for addr in address_list:
    jieba.add_word(addr)
```

```python
all_address_pd = pd.read_csv('../ext_data/nbs_id_card_code_ver1.csv')
```

```python
all_address_pd['address'] = all_address_pd.loc[:,'address'].apply(lambda x : remove_unuse_char(x))
```

```python
all_list =  all_address_pd.loc[:,'address']
```

```python
all_doc_list = []
for addr in all_list:
#     doc_list = [word for word in jieba.cut_for_search(addr)] #best
#     doc_list = [word for word in jieba.cut(addr)]
    doc_list = [word for word in jieba.cut(addr,cut_all=True)]
    all_doc_list.append(doc_list)
```

```python
dictionary = corpora.Dictionary(all_doc_list)
```

```python
corpus = [dictionary.doc2bow(doc) for doc in all_doc_list]
```

```python
tfidf = models.TfidfModel(corpus)
```

```python
index = similarities.SparseMatrixSimilarity(tfidf[corpus], num_features=len(dictionary.keys()))
```

# 纠错开始

```python
from Levenshtein import distance
```

```python
correct_pd = result.copy()
cannot_correct = []
for idx,error_id in enumerate(error_file_id):
#     print(error_id[0])
#     print(result.loc[result.file_id == error_id[0], 'addr'].values[0])

    doc_test= result.loc[result.file_id == error_id[0], 'addr'].values[0]
    
    if len(doc_test) == 0:
        print(error_id[0] +" is nan")
        correct_pd.loc[correct_pd.file_id == error_id[0], 'addr'] = ''
        continue
    
#     doc_test_list = [word for word in jieba.cut_for_search(doc_test)] # best
    doc_test_list = [word for word in jieba.cut(doc_test,cut_all=True)] 
    
    doc_test_vec = dictionary.doc2bow(doc_test_list)
    sim = index[tfidf[doc_test_vec]]
    
    min_sim_dist = 999
    most_sim_addr = ""
    most_prob = 0
#     for sim_idx in sorted(enumerate(sim), key=lambda item: -item[1])[0:50]:
#     for sim_idx in sorted(enumerate(sim), key=lambda item: -item[1])[0:150]: #best
    for sim_idx in sorted(enumerate(sim), key=lambda item: -item[1])[0:200]: #best
        sim_addr = remove_unuse_char(all_address_pd.iloc[sim_idx[0],12])
        sim_ed_dist = distance(doc_test,sim_addr)
        
        if sim_ed_dist < min_sim_dist:
            min_sim_dist = sim_ed_dist
            most_sim_addr = sim_addr
            most_prob = sim_idx[1]
            
    
    if (most_prob < 0.5 and min_sim_dist / len(doc_test) > 0.5):
        print('error! ',idx,min_sim_dist,most_prob,error_id[0],doc_test,'-->>',most_sim_addr)
        cannot_correct.append([error_id[0]])
        most_sim_addr = doc_test
    else:
        print(idx,min_sim_dist,most_prob,error_id[0],doc_test,'-->>',most_sim_addr)
    
    
    correct_pd.loc[correct_pd.file_id == error_id[0], 'addr'] = most_sim_addr
        
    
#     new_addr = remove_unuse_char(all_address_pd.iloc[similarity_index,12])
    
#     print(idx,error_id[0],doc_test,'-->>',new_addr)
#     correct_pd.loc[correct_pd.file_id == error_id[0], 'addr'] = new_addr
#     break
```
